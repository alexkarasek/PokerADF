```sql
-- Create an external file format for PARQUET files.  
CREATE EXTERNAL FILE FORMAT parquetfile1  
WITH (  
    FORMAT_TYPE = PARQUET,  
    DATA_COMPRESSION = 'org.apache.hadoop.io.compress.SnappyCodec'  
); 
```

```sql
--Create External tables to faciliate queries
IF OBJECT_ID('stg.HH_External') IS NOT NULL
DROP EXTERNAL TABLE stg.HH_External

CREATE EXTERNAL TABLE [stg].[HH_External] (
    id varchar(20),
    [timestamp] varchar(50),
    sitename varchar(50),
    tablename varchar(100),
    limits varchar(50),
    buttonseat bigint,
    gameactionid bigint,
    streetname varchar(100),
    action varchar(20),
    actor varchar(100),
    amount float
)
WITH
(
    LOCATION='/Data/Test/Struct/Sample'
,   DATA_SOURCE = AzureDataLakeStorage
,   FILE_FORMAT = parquetfile1
,   REJECT_TYPE = VALUE
,   REJECT_VALUE = 0
)
;
```

```sql
select count(distinct [id])
from stg.HH_External

select [id], sum(cast(amount as [decimal](9,2)))
from stg.HH_External
group by [id]
having sum(cast(amount as [decimal](9,2))) <> 0

select *
from stg.HH_External
order by [id], gameactionid

```
